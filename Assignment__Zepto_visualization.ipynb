```python
# Load the dataset
file_path = '/content/drive/MyDrive/IIT BHU - DS Assessment dataset/data.csv'
data = pd.read_csv(file_path)
```

```python
#Display the infor of the dataset
data.info()
```

```python
# Display the first few rows of the dataset
data.head(), data.columns
```

```python
# Set the size of the figure
plt.figure(figsize=(16, 12))

# Select only numerical columns for correlation calculation
numerical_data = data.select_dtypes(include=['float', 'int'])

# Generate a correlation matrix
corr_matrix = numerical_data.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)

# Set the title
plt.title('Correlation Heatmap of Numerical Features')

# Show the plot
plt.show()
```

```python
# Plotting distribution of search_term (top 20 for readability)
plt.figure(figsize=(8, 4))
top_search_terms = data['search_term'].value_counts().nlargest(20).index
sns.countplot(y='search_term', data=data[data['search_term'].isin(top_search_terms)], order=top_search_terms)
plt.title('Distribution of Top 20 Search Terms')
plt.xlabel('Count')
plt.ylabel('Search Term')
plt.show()
```

```python
# Plot 1: Most Frequently Searched Terms
plt.figure(figsize=(8, 4))
top_search_terms = data['search_term'].value_counts().head(10)
sns.barplot(x=top_search_terms.values, y=top_search_terms.index, palette='viridis')
plt.title('Top 10 Most Frequently Searched Terms')
plt.xlabel('Frequency')
plt.ylabel('Search Term')
plt.show()
```

```python
# Plot 2: CTR Trends Over the Last 30 and 7 Days
plt.figure(figsize=(8, 4))
```

```python

plt.figure(figsize=(20,9))  # Adjust the width and height as needed
plt.subplot(1, 2, 1)
sns.lineplot(data=data, x='search_term', y='CTR_last_30_days', marker='o')
plt.xticks(rotation=90)
plt.title('CTR Last 30 Days')
plt.xlabel('Search Term')
plt.ylabel('CTR')
plt.show()
```

```python
# Last 7 days
plt.figure(figsize=(20,9))  # Adjust the width and height as needed
plt.subplot(1, 2, 2)
sns.lineplot(data=data, x='search_term', y='CTR_last_7_days', marker='o', color='orange')
plt.xticks(rotation=90)
plt.title('CTR Last 7 Days')
plt.xlabel('Search Term')
plt.ylabel('CTR')

plt.tight_layout()
plt.show()
```

```python
# Plot 3: Relationship Between Clicks and Ad Revenue
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='total_clicks', y='ad_revenue', hue='query_type', palette='deep')
plt.title('Relationship Between Clicks and Ad Revenue')
plt.xlabel('Total Clicks')
plt.ylabel('Ad Revenue')
plt.legend(title='Query Type')
plt.show()
```
